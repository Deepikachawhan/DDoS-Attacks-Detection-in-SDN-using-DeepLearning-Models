```python
import seaborn as sns
import pandas as pd
import numpy as np


# Load your dataset
df=pd.read_csv("/content/drive/MyDrive/Minor_project6thsem/Datasets/DDOS SDN TREE TORUS /DDoS SDN Tree Torus ")

# Print the number of columns in the original dataset
print("Original number of columns:", len(df.columns))
print(df.columns)

# Select only numeric columns
numeric_df = df.select_dtypes(include=[np.number])
print("Number of numeric columns:", len(numeric_df.columns))
print(numeric_df.columns)

# Calculate the correlation matrix
corr = numeric_df.corr()
correlation_with_label = corr['label'].sort_values(ascending=False)
print("Correlation with 'label':\n", correlation_with_label)

# Drop features with correlation < 0.1 (absolute value)
low_corr_features = correlation_with_label[abs(correlation_with_label) < 0.1].index
filtered_numeric_df = numeric_df.drop(columns=low_corr_features)

# Print shape after dropping low-correlation features
print(f"Dropped features: {list(low_corr_features)}")
print(f"Remaining features: {list(filtered_numeric_df.columns)}")
print(f"Number of remaining features: {len(filtered_numeric_df.columns)}")

# Re-calculate the correlation matrix with filtered features
filtered_corr = filtered_numeric_df.corr()

# Plot heatmap for remaining features
plt.figure(figsize=(12, 8))
sns.heatmap(filtered_corr, annot=True, fmt='.2f', cmap='coolwarm')
plt.title("Correlation Matrix (Filtered Features)")
plt.show()
```

Random Forest Classifier


```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay, accuracy_score, roc_curve, auc
import matplotlib.pyplot as plt
```

```python
# Drop non-numeric and time-based features
non_numeric_columns = ['src', 'dst', 'Protocol', 'eth.src', 'eth.dst', 'ip.proto']
time_features = ['interval_start', 'frame.time_epoch', 'frame.time_delta']
df = df.drop(columns=non_numeric_columns + time_features)
```

```python
# Separate features and target
X = df.drop(columns=['label'])
y = df['label']
```

```python
# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42, stratify=y)
```

```python
# Train Random Forest with regularization
rf_model = RandomForestClassifier(
    n_estimators=100,
    max_depth=10,               # Limit tree depth
    min_samples_split=10,        # Min samples to split a node
    random_state=42
)
rf_model.fit(X_train, y_train)
```

```python
# Predictions
y_pred = rf_model.predict(X_test)
```

```python
# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Normal (0)', 'Attack (1)'])
disp.plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix (Time Features Dropped)')
plt.show()
```

```python
# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy (Time Features Dropped): {accuracy:.4f}')
```

```python
# Classification Report
print('Classification Report (Time Features Dropped):')
print(classification_report(y_test, y_pred))
```

```python
# ROC Curve
y_prob = rf_model.predict_proba(X_test)[:, 1]
fpr, tpr, _ = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)
```

```python
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()
```

```python
# Feature Importances
importances = rf_model.feature_importances_
feature_names = X.columns
indices = np.argsort(importances)[::-1]
```

```python
plt.figure(figsize=(12, 8))
plt.title('Feature Importances (Random Forest)')
plt.bar(range(X.shape[1]), importances[indices], align='center')
plt.xticks(range(X.shape[1]), feature_names[indices], rotation=90)
plt.xlabel('Features')
plt.ylabel('Importance Score')
plt.show()
```

```python
# Accuracy and loss over iterations (simulating with oob_score for Random Forest)
n_estimators = list(range(1, 101, 10))
oob_scores = []

for n in n_estimators:
    rf = RandomForestClassifier(n_estimators=n, max_depth=10, min_samples_split=10, oob_score=True, random_state=42, bootstrap=True)
    rf.fit(X_train, y_train)
    oob_scores.append(1 - rf.oob_score_)

plt.figure(figsize=(10, 6))
plt.plot(n_estimators, oob_scores, marker='o')
plt.title('Out-of-Bag Error Rate (Proxy for Loss)')
plt.xlabel('Number of Trees (n_estimators)')
plt.ylabel('OOB Error Rate (Loss)')
plt.grid(True)
plt.show()
```

```python
# Accuracy plot
train_accuracies = []
test_accuracies = []
```

```python
for n in n_estimators:
    rf = RandomForestClassifier(n_estimators=n, max_depth=10, min_samples_split=10, random_state=42)
    rf.fit(X_train, y_train)
    train_accuracies.append(rf.score(X_train, y_train))
    test_accuracies.append(rf.score(X_test, y_test))
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, LSTM, Conv1D, MaxPooling1D, Flatten
```

```python
# Define a simple MLP model
def create_mlp():
    model = Sequential([
        Dense(128, activation='relu', input_shape=(X_train.shape[1],)),
        Dropout(0.3),
        Dense(64, activation='relu'),
        Dropout(0.3),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model
```

```python
# Define a simple CNN model
def create_cnn():
    model = Sequential([
        Conv1D(64, kernel_size=3, activation='relu', input_shape=(X_train.shape[1], 1)),
        MaxPooling1D(pool_size=2),
        Flatten(),
        Dense(64, activation='relu'),
        Dropout(0.3),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model
```

```python
# Define a simple LSTM model
def create_lstm():
    model = Sequential([
        LSTM(64, input_shape=(X_train.shape[1], 1)),
        Dropout(0.3),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model
```

```python
# Reshape for CNN and LSTM
X_train_cnn = X_train[..., np.newaxis]
X_test_cnn = X_test[..., np.newaxis]
```

```python
# Train and evaluate MLP
mlp_model = create_mlp()
mlp_history = mlp_model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=10, batch_size=32)
```

```python
# Train and evaluate CNN
cnn_model = create_cnn()
cnn_history = cnn_model.fit(X_train_cnn, y_train, validation_data=(X_test_cnn, y_test), epochs=10, batch_size=32)
```

```python
# Train and evaluate LSTM
lstm_model = create_lstm()
lstm_history = lstm_model.fit(X_train_cnn, y_train, validation_data=(X_test_cnn, y_test), epochs=10, batch_size=32)
```

```python
# Evaluate models
def plot_history(history, title):
    plt.figure(figsize=(12, 6))
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title(f'{title} Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
plot_history(mlp_history, 'MLP')

```

```python
plot_history(cnn_history, 'CNN')
```

```python
plot_history(lstm_history, 'LSTM')
```

```python
# Predictions and confusion matrices
def evaluate_model(model, X_test, y_test, model_name):
    y_pred = (model.predict(X_test) > 0.5).astype('int32')
    cm = confusion_matrix(y_test, y_pred)
    print(f'Confusion Matrix for {model_name}:\n', cm)
    print(f'Classification Report for {model_name}:\n', classification_report(y_test, y_pred))
    print(f'Accuracy for {model_name}: {accuracy_score(y_test, y_pred):.4f}\n')

     # Plot confusion matrix
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Normal', 'Attack'], yticklabels=['Normal', 'Attack'])
    plt.xlabel('Predicted Labels')
    plt.ylabel('True Labels')
    plt.title(f'Confusion Matrix for {model_name}')
    plt.show()
```

```python
evaluate_model(mlp_model, X_test, y_test, 'MLP')
```

```python
evaluate_model(cnn_model, X_test_cnn, y_test, 'CNN')
```

```python
evaluate_model(lstm_model, X_test_cnn, y_test, 'LSTM')
```
